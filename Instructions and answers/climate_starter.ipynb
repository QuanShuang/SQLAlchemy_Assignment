# Reflect Tables into SQLAlchemy ORM

```python
# Python SQL toolkit and Object Relational Mapper
import sqlalchemy
from sqlalchemy.ext.automap import automap_base
from sqlalchemy.orm import Session
from sqlalchemy import create_engine, func, inspect
```

```python
engine = create_engine("sqlite:///Resources/hawaii.sqlite")
```

```python
# reflect an existing database into a new model
Base = automap_base()
# reflect the tables
Base.prepare(engine, reflect=True)
```

```python
# We can view all of the classes that automap found
Base.classes.keys()
```

```python
# Save references to each table
Measurement = Base.classes.measurement
Station = Base.classes.station
```

```python
# Create our session (link) from Python to the DB
session = Session(engine)
```

## Temperature Analysis

```python
# This function called `calc_temps` will accept start date and end date in the format '%Y-%m-%d' 
# and return the minimum, average, and maximum temperatures for that range of dates
def calc_temps(start_date, end_date):
    """TMIN, TAVG, and TMAX for a list of dates.
    
    Args:
        start_date (string): A date string in the format %Y-%m-%d
        end_date (string): A date string in the format %Y-%m-%d
        
    Returns:
        TMIN, TAVE, and TMAX
    """
    
    return session.query(func.min(Measurement.tobs), func.avg(Measurement.tobs), func.max(Measurement.tobs)).\
        filter(Measurement.date >= start_date).filter(Measurement.date <= end_date).all()

# function usage example
print(calc_temps('2012-02-28', '2012-03-05'))
```

```python
# Use your previous function `calc_temps` to calculate the tmin, tavg, and tmax 
# for your trip using the previous year's data for those same dates.
# Date of trip: ('2017-04-01','2017-04-10')
start_date = '2017-04-01'
end_date='2017-04-10'
start_date_yearago = dt.datetime.strptime(start_date, '%Y-%m-%d')-dt.timedelta(days=365)
end_date_yearago = dt.datetime.strptime(end_date, '%Y-%m-%d')-dt.timedelta(days=365)

Tmin=calc_temps(start_date_yearago,end_date_yearago)[0][0]
Tmax=calc_temps(start_date_yearago,end_date_yearago)[0][2]
Tavg=calc_temps(start_date_yearago,end_date_yearago)[0][1]
print(f"During the trip, max temp was {Tmax}, min temp was {Tmin} and average temp was {Tavg} (Degree F).")
```

```python
# Plot the results from your previous query as a bar chart. 
# Use "Trip Avg Temp" as your Title
# Use the average temperature for the y value
# Use the peak-to-peak (tmax-tmin) value as the y error bar (yerr)
df = pd.DataFrame({"avg":[Tavg],"error":[(Tmax-Tmin)]})
means = df["avg"]
sem = df["error"]
plt.figure(figsize=(4,8))
plot=means.plot(kind='bar',yerr=sem,color = "lightsalmon")
plt.xticks([])
plt.xlabel("")
plt.ylim(0,100)
plt.ylabel("Temp (F)")
plt.title("Trip Avg Temp")

plt.show()
```

```python
# Calculate the total amount of rainfall per weather station for your trip dates using the previous year's matching dates.
# Sort this in descending order by precipitation amount and list the station, name, latitude, longitude, and elevation
session.query(Measurement.station,Station.name,Station.latitude,Station.longitude,Station.elevation,func.sum(Measurement.prcp),).\
filter(Measurement.date >= start_date_yearago).filter(Measurement.date <= end_date_yearago).group_by(Measurement.station).\
order_by(func.sum(Measurement.prcp).desc()).all()
```

## Daily Rainfall Average

```python
# Create a query that will calculate the daily normals 
# (i.e. the averages for tmin, tmax, and tavg for all historic data matching a specific month and day)

def daily_normals(date):
    """Daily Normals.
    
    Args:
        date (str): A date string in the format '%m-%d'
        
    Returns:
        A list of tuples containing the daily normals, tmin, tavg, and tmax
    
    """
    
    sel = [func.min(Measurement.tobs), func.avg(Measurement.tobs), func.max(Measurement.tobs)]
    return session.query(*sel).filter(func.strftime("%m-%d", Measurement.date) == date).all()
    
daily_normals("01-01")
```

```python
# Set the start and end date of the trip
print(start_date, end_date)
# Use the start and end date to create a range of dates
s = dt.datetime.strptime(start_date, '%Y-%m-%d')
e = dt.datetime.strptime(end_date, '%Y-%m-%d')

ds=date(s.year,s.month,s.day)
de=date(e.year,e.month,e.day)

delta=de-ds 

# Stip off the year and save a list of %m-%d strings
list_dates=[]
for i in range(delta.days +1):
    list_dates.append(str(ds + timedelta(i))[5:])

print(list_dates)
```

```python
# Loop through the list of %m-%d strings and calculate the normals for each date
normals=[]
for date in list_dates:
    normals.append(daily_normals(date))
normals
```

```python
# loop and put Tmin, Tmax, Tavg into lists
tmin_list=[]
tmax_list=[]
tavg_list=[]
for j in range(len(normals)):
    tmin_list.append(normals[j][0][0])
    tmax_list.append(normals[j][0][2])
    tavg_list.append(normals[j][0][1])
```

```python
# since the date we used is month+date, now loop again to restore year into date lists
list_dates_withyear=[]
for i in range(delta.days +1):
    list_dates_withyear.append(str(ds + timedelta(i)))

print(list_dates_withyear)
```

```python
# Load the previous query results into a Pandas DataFrame and add the `trip_dates` range as the `date` index
rain=pd.DataFrame({"Date":list_dates_withyear,"Tmin":tmin_list,"Tavg":tavg_list,"Tmax":tmax_list}).set_index("Date")
rain
```

```python
# Plot the daily normals as an area plot with `stacked=False`
rain.plot(kind='area', stacked=False, alpha=0.5,title='Daily Rainfall Average')
plt.legend(loc='lower left')
plt.xticks(np.arange(9),list_dates_withyear,rotation = '45')
plt.show()
```
